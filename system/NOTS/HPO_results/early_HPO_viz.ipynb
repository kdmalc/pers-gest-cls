```python
import optuna
from optuna.storages import JournalStorage
from optuna.storages.journal import JournalFileBackend, JournalFileOpenLock
import optuna.visualization as vis
```

## CURRENT 100 TRIAL LOG

```python
# 1. Define the path to your log file
log_path = r"C:\\Users\\kdmen\\Repos\\pers-gest-cls\\dataset\\optuna_dbs\\mamlpp_CNNLSTMMLP_2fcv_hpo.log"
```

```python
# 2. Setup Journal Storage with a file lock (crucial for Windows)
lock_obj = JournalFileOpenLock(log_path)
storage = JournalStorage(JournalFileBackend(log_path, lock_obj=lock_obj))
```

```python
# 3. Load the study
# Note: Since journal logs can store multiple studies, we fetch all names 
# and load the first one (or you can specify the name if you know it).
# Use the updated function to list studies
summaries = optuna.study.get_all_study_summaries(storage=storage)

if not summaries:
    print("No studies found in the log file. Check your file path or if the HPO actually ran.")
else:
    # Get the name of the first study found
    target_study_name = summaries[0].study_name
    
    # Load the study
    study = optuna.load_study(study_name=target_study_name, storage=storage)
    
    print(f"Successfully loaded: {target_study_name}")
    print(f"Number of finished trials: {len(study.trials)}")
```

```python
# --- VISUALIZATIONS ---

# A. Performance across trials (Optimization History)
# This shows how validation accuracy improved over the course of the HPO
fig_history = vis.plot_optimization_history(study)
fig_history.show()
```

```python
# B. Hyperparameter Interactions (Contour Plot)
# This specifically shows how pairs of hyperparameters interact relative to accuracy
#fig_contour = vis.plot_contour(study)
#fig_contour.show()
```

```python
# C. Parallel Coordinate Plot
# Great for seeing high-dimensional relationships and which paths lead to high accuracy
#fig_parallel = vis.plot_parallel_coordinate(study)
#fig_parallel.show()
```

```python
# D. Parameter Importance
# Helps you identify which hyperparameters actually moved the needle
fig_importance = vis.plot_param_importances(study)
fig_importance.show()
```

```python
# Calculate parameter importance first
importance = optuna.importance.get_param_importances(study)

# Get the names of the top 8 most influential hyperparameters
top_params = list(importance.keys())[:8]

# Now, only plot those 8 in your interaction plots
## Still too much going on in these plots...
#fig_contour = vis.plot_contour(study, params=top_params)
#fig_contour.show()

fig_parallel = vis.plot_parallel_coordinate(study, params=top_params)
fig_parallel.show()
```
